```python
from layers import Linear
import numpy as np
X = np.random.randn(3,2) 
Weight = np.random.randn(2,5)*0.01
bias = np.zeros((1,5))

linear = Linear(in_dims = 2 ,neurons_num =  5) 
linear.W = Weight
linear.b = bias
Z = linear.forward(X) 
dZ = np.ones_like(Z)
print(Z.shape) 
dX = linear.backward(dZ)
print("shape của dX và X bằng nhau là: ", dX.shape == X.shape)
print("shape của db và b bằng nhau là: ", linear.db.shape == bias.shape)
print("shape của dW và W bằng nhau là: ", linear.dW.shape == Weight.shape)

```
