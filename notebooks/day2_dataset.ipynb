```python
import numpy as np

def generate_circle_data(n=2000, r=0.5, seed=42):
    np.random.seed(seed)
    
    X = np.random.uniform(-1, 1, size=(n, 2))
    y = (X[:, 0]**2 + X[:, 1]**2 < r**2).astype(int)
    
    return X, y
import numpy as np
import matplotlib.pyplot as plt

from data.generate import generate_circle_data

X, y = generate_circle_data(n=2000, r=0.5, seed=42)

X.shape, y.shape

np.random.seed(42)
idx = np.random.permutation(len(X))
train_idx = idx[:1600]
test_idx = idx[1600:]
X_train , y_train = X[train_idx] ,y[train_idx] 
X_test,  y_test  = X[test_idx],  y[test_idx]
plt.figure(figsize=(4,4))
plt.scatter(X[:,0], X[:,1], c=y, cmap="coolwarm", s=10)
plt.gca().set_aspect("equal")
plt.title("Circle classification dataset")
plt.show()
```
