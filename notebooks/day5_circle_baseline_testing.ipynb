```python
from data.generate import generate_circle_data 
from layers import Linear 
from activations import ReLU , Sigmoid
from losses import BCE 
from train_loop import train 
import numpy as np

np.set_printoptions(suppress=True) 
model = [Linear(2,16),ReLU(),Linear(16,1),Sigmoid()] 

X,y =  generate_circle_data(2000,0.5,42)
loss_func = BCE()
train(model,loss_func,X,y,1000,32,0.1)



def Predict(X,model): 
    A = X 
    for layer in model: 
        A = layer.forward(A) 
    return A

y_hat = Predict(X,model) 
threshold = 0.5
y_pred = y_hat.copy()

shuffle = np.random.permutation(len(y_pred))
y_pred = y_pred[shuffle]
y = y[shuffle]

y_pred = y_pred[0:100]
y = y[0:100]

for i in range(100): 
    if y_pred[i] >= threshold: 
        y_pred[i] = 1
    else:
        y_pred[i] = 0


accuracy = 100
for i in range(100):
    
    if y_pred[i] - y[i] != 0:
        accuracy -=1

print(accuracy, "%") 
        
```
